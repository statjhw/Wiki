```python
import sqlite3

# DB 파일 연결 (파일이 없으면 생성됨)
conn = sqlite3.connect('sql_ex.db')

# 커서 생성 (SQL 실행을 위한 도구)
cursor = conn.cursor()

# 예시: 테이블 조회
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())

# 연결 종료
conn.close()
```

```python
class DB:
    def __init__(self):
        self.conn = sqlite3.connect('sql_ex.db')
        self.cursor = self.conn.cursor()

    def excute(self, sql):
        self.cursor.execute(sql)
        for row in self.cursor.fetchall():
            print(row)

    def close(self):
        self.conn.close()
```

```python
db = DB()
```
